```python
img = cv2.imread('..\\data\\boy.jpg')
copy = np.copy(img)
```

```python
img.shape
```

```python
mask = np.zeros(img.shape[:2],dtype=np.uint8)
```

```python
mask.shape
```

```python
#create dummy array
bgd = np.zeros((1,65),np.float64)
fgd = np.zeros((1,65),np.float64)
```

```python
# Select region of interest 
```

```python
# rectangle = cv2.selectROI(copy)
x,y,w,h = rect= (41, 130, 915, 427)
```

```python
m,s,e =cv2.grabCut(copy,mask,rect,bgd,fgd,5,cv2.GC_INIT_WITH_RECT)
```

```python
np.unique(m)
```

```python
_= np.copy(img)
cv2.rectangle(_,(x,y),(x+w,h+y),(50,40,200),3)
plt.imshow(_[:,:,::-1])
plt.axis('off')
```

```python
mask2 = np.where((m == 0) | (m == 2),0,1).astype(np.uint8)
plt.imshow(m)
plt.axis('off')
```

```python
plt.imshow(mask2)
plt.axis('off')
```

```python
# Inference on original image
img = img * mask2[:,:,np.newaxis]
```
